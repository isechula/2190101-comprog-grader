```python
# 06_Func_NextPrime

def is_prime(n):
    if n <= 1:
        return False
    for k in range(2, int(n**0.5)+1):
        if n%k == 0:
            return False
    return True

def next_prime(n):
    i = 1
    while True:
        if is_prime(n+i):
            break
        else:
            i += 1
    return n+i

def next_twin_prime(n):
    i = 1
    while True:
        if is_prime(n+i) and is_prime(n+i+2):
            break
        else:
            i += 1
    return (n+i, n+i+2)

exec(input().strip())
```

```python
# 06_Func_Function_call

def read_answers():
    N = int(input())
    answers = []
    for _ in range(N):
        sid, ans = input().split()
        answers.append([sid, ans])
    return answers

def scoring(answers, solution):
    scores = []
    for sid, ans in answers:
        score = marking(ans, solution) / len(solution) * 100
        grade = grading(score)
        scores.append([sid, score, grade])
    return scores

def marking(answer, solution):
    score = 0
    for i in range(len(answer)):
        if answer[i] == solution[i]:
            score += 1
    return score

def grading(score):
    g = [[80, "A"], [70, "B"], [60, "C"], [50, "D"]]
    for a, b in g:
        if score >= a:
            return b
    return "F"

def report(scores):
    for sid, sc, grade in scores:
        print(sid, sc, grade)

def sort(scores):
    x = []
    for sid, score, grade in scores:
        x.append([score, sid, grade])
    x.sort(reverse=True)
    for i in range(len(x)):
        scores[i] = [x[i][1], x[i][0], x[i][2]]
    return scores


solution = input()
answers = read_answers()
report(sort(scoring(answers, solution)))
```

```python
# 06_Func_Distance

import math

def distance1(x1, y1, x2, y2):
    difx = abs(float(x1)-float(x2))
    dify = abs(float(y1)-float(y2))
    return math.sqrt(difx**2+dify**2)

def distance2(p1, p2):
    difx = abs(float(p1[0])-float(p2[0]))
    dify = abs(float(p1[1])-float(p2[1]))
    return math.sqrt(difx**2+dify**2)

def distance3(c1, c2):
    dx = c2[0] - c1[0]
    dy = c2[1] - c1[1]
    center_distance = math.sqrt(dx**2 + dy**2)
    sum_of_radii = c1[2] + c2[2]
    overlap = center_distance <= sum_of_radii
    return center_distance, overlap

def perimeter(points):
    total_distance = 0
    n = len(points)
    for i in range(n):
        current_point = points[i]
        next_point = points[(i + 1) % n]
        
        dx = next_point[0] - current_point[0]
        dy = next_point[1] - current_point[1]
        distance = math.sqrt(dx**2 + dy**2)
        total_distance += distance
    return total_distance

exec(input().strip())
```

```python
# 06_Func_FourFunctions

def make_int_list(x):
    return [int(i) for i in x.split()]

def is_odd(e):
    return e%2!=0

def odd_list(alist):
    return [x for x in alist if x%2!=0]

def sum_square(alist):
    return sum([x**2 for x in alist])

exec(input().strip())
```
